#### 2. La Serie de Pandas (`Series`)

La **Serie** es el componente básico de Pandas. Es una estructura de datos unidimensional, como una columna en una hoja de cálculo.

2.  **Índice:** Tiene etiquetas para cada elemento (por defecto, números que empiezan en 0).

**Ejemplo:** Crearemos una Serie con las alturas de 5 personas.

```python
import pandas as pd

# Datos de ejemplo: Alturas en centímetros
alturas = [175, 182, 165, 178, 190]

# Crear la Serie de Pandas
serie_alturas = pd.Series(alturas)

print("--- La Serie de Alturas ---")
print(serie_alturas)

print("\nTipo de objeto creado:")
print(type(serie_alturas))
```

# 📝 Explicación Completa del Método `.iloc` en Pandas

El método **`.iloc`** (acrónimo de *integer location* o **localización por índice entero**) es la herramienta principal de la biblioteca Pandas para seleccionar datos de un DataFrame **únicamente basándose en su posición numérica** (su índice).

A diferencia de `.loc`, que usa los nombres de las filas y columnas, `.iloc` **ignora** las etiquetas y trabaja estrictamente con números enteros, comenzando el conteo en **cero (0)**.

### 📐 Sintaxis Básica

La sintaxis siempre sigue el formato:

$$\text{df.iloc} [ \text{filas}, \text{columnas} ]$$

Donde la selección para `filas` y `columnas` puede ser:

| Tipo de Selección | Descripción | Ejemplo en Código | Resultado |
| :--- | :--- | :--- | :--- |
| **Entero Único** | Selecciona la fila o columna en esa posición exacta (índice 0, 1, 2...). | `df.iloc[0]` | La primera fila completa. |
| **Lista de Enteros** | Selecciona múltiples filas o columnas no contiguas. | `df.iloc[[0, 5, 9], [1, 3]]` | Filas en las posiciones 0, 5, 9 y Columnas en 1 y 3. |
| **Rebanada (*Slicing*)** | Selecciona un rango contiguo de filas o columnas. **¡Cuidado!** El índice de fin es **exclusivo**. | `df.iloc[2:5]` | Filas en las posiciones 2, 3 y 4 (omite la posición 5). |
| **Selección Completa** | Se usa el carácter de dos puntos (`:`) para seleccionar todos los elementos de esa dimensión. | `df.iloc[:, 2]` | Todas las filas de la columna en la posición 2. |

### 🧭 Ejemplos Clave de Uso

| Objetivo | Código de `.iloc` | Lo que Selecciona |
| :--- | :--- | :--- |
| **Una Fila** | `df.iloc[1]` | La segunda fila completa. |
| **Una Columna** | `df.iloc[:, 3]` | La cuarta columna completa. |
| **Un Valor Único** | `df.iloc[2, 4]` | El valor ubicado en la tercera fila y quinta columna. |
| **Primeras N Filas** | `df.iloc[:5]` | Las primeras 5 filas (índices 0 a 4). |
| **Últimas N Filas** | `df.iloc[-5:]` | Las últimas 5 filas (utiliza índices negativos, como en Python). |
| **Subconjunto Rectangular** | `df.iloc[0:10, 5:8]` | Filas del 0 al 9 y columnas del 5 al 7. |

La indexación con `.iloc` es **absolutamente posicional**. Es el método más rápido para acceder a datos por posición.

# 🚀 Funcionalidades Avanzadas del Método `.iloc`

Más allá de la simple selección, `.iloc` es una herramienta poderosa para modificar y acceder a elementos específicos del DataFrame usando la posición.

### 1. Acceso con Índices Negativos (Reverse Indexing)

Al igual que en las listas de Python, los índices negativos en `.iloc` permiten contar las filas o columnas desde el final del DataFrame:

* **`-1`**: Hace referencia al **último** elemento (última fila o última columna).
* **`-2`**: Hace referencia al penúltimo elemento, y así sucesivamente.

| Objetivo | Código de `.iloc` | Lo que Selecciona |
| :--- | :--- | :--- |
| **Última Fila** | `df.iloc[-1]` | La última fila completa del DataFrame. |
| **Última Columna** | `df.iloc[:, -1]` | La última columna completa (como Series). |
| **Penúltima a Última** | `df.iloc[-2:]` | Las dos últimas filas. |

### 2. Modificación y Asignación de Valores

`.iloc` es fundamental para **modificar los datos** de manera posicional. Se utiliza el operador de asignación (`=`) después de seleccionar el subconjunto deseado.

| Tarea de Modificación | Código de Asignación |
| :--- | :--- |
| **Cambiar un Valor Único** | `df.iloc[3, 0] = 9999` |
| **Cambiar una Fila Completa** | `df.iloc[0] = [100, 200, ...]` |
| **Cambiar un Rango de Celdas** | `df.iloc[5:7, 1:4] = 0` |

> **Nota:** Al asignar valores, el valor único o la longitud del *array*/lista deben coincidir con la forma del subconjunto seleccionado para evitar errores.

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.
try:
    df = df_california.head(8).copy()
    # Reducimos las columnas para que sea fácil de visualizar la modificación
    df = df.iloc[:, 0:4]
except NameError:
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrame de prueba.")
    data = np.arange(1, 33).reshape(8, 4)
    df = pd.DataFrame(data, columns=['Col_A', 'Col_B', 'Col_C', 'Col_D'])

print("--- DataFrame Inicial (8 filas, 4 columnas) ---")
print(df)
print("\n" + "="*50 + "\n")


# === 1. Acceso con Índices Negativos ===

# a) Seleccionar la última fila (-1)
print("--- 1a. Última Fila (iloc[-1]) ---")
ultima_fila = df.iloc[-1]
print(ultima_fila)
print("\n")


# b) Seleccionar la última columna para todas las filas
print("--- 1b. Última Columna (iloc[:, -1]) ---")
ultima_columna = df.iloc[:, -1]
print(ultima_columna)
print("\n")


# === 2. Modificación y Asignación de Valores ===

# a) Modificar un valor único: Fila 0 (primera), Columna 0 (primera)
print("--- 2a. Modificando un valor único: [0, 0] a 1000 ---")
df.iloc[:,0] = 1000
print(f"Nuevo valor en [0, 0]: {df.iloc[0, 0]}")
print("\n")


# b) Modificar un rango de valores a 0: Filas 2 y 3 (2:4), Columnas 1 y 2 (1:3)
print("--- 2b. Modificando rango: Filas 2:4, Cols 1:3 a 0 ---")
df.iloc[2:4, 1:3] = 0
print("DataFrame después de la modificación del rango:")
print(df)
```

# 🎯 Último Caso de Uso de `.iloc`: Selección de Columnas como DataFrame vs. Series

Cuando seleccionamos una columna o un conjunto de columnas, el resultado que devuelve `.iloc` puede ser diferente (una **Serie** o un **DataFrame**), lo cual tiene implicaciones sobre cómo puedes seguir manipulando esos datos.

### 1. Obtener una Columna Única como una *Serie* (El Comportamiento Estándar)

Si usas la indexación simple, Pandas devolverá el resultado como una `Series` (una matriz unidimensional etiquetada).

* **Sintaxis:** `df.iloc[:, índice_columna]`
* **Resultado:** Objeto Pandas `Series`.
* **Ejemplo:** `df.iloc[:, 0]`

### 2. Obtener una Columna Única como un *DataFrame*

A veces, necesitas que la columna seleccionada se mantenga como un DataFrame (una matriz bidimensional). Esto es crucial si planeas aplicar transformaciones que solo funcionan en DataFrames o si quieres mantener los nombres de las columnas explícitos.

Para forzar que el resultado sea un DataFrame, debes pasar la posición de la columna como una **lista de índices** (incluso si solo hay uno).

* **Sintaxis:** `df.iloc[:, [índice_columna]]` (Nótese el doble corchete `[]`)
* **Resultado:** Objeto Pandas `DataFrame`.
* **Ejemplo:** `df.iloc[:, [0]]`

### Resumen Visual

| Código `.iloc` | Forma del Resultado | Tipo de Objeto |
| :--- | :--- | :--- |
| `df.iloc[:, 0]` | (n,) | `Series` |
| `df.iloc[:, [0]]` | (n, 1) | `DataFrame` |

Este es el último detalle clave de la sintaxis de `.iloc`.

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.
try:
    df = df_california.head(5).copy()
except NameError:
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrame de prueba.")
    data = np.arange(1, 16).reshape(5, 3)
    df = pd.DataFrame(data, columns=['Median_Income', 'Housing_Median_Age', 'Total_Rooms'])


print("--- DataFrame Inicial (Primeras 5 filas) ---")
print(df)
print("\n" + "="*50 + "\n")


# 1. Seleccionar la primera columna (índice 0) como una SERIE (Series)
# Usando la sintaxis simple: df.iloc[:, indice]
columna_series = df.iloc[:, 0]
print("--- 1. Columna 0 como SERIE ---")
print(columna_series)
print(f"Tipo de Objeto: {type(columna_series)}")
print(f"Forma (Shape): {columna_series.shape}")
print("\n")


# 2. Seleccionar la primera columna (índice 0) como un DATAFRAME
# Usando la sintaxis de lista: df.iloc[:, [indice]]
columna_dataframe = df.iloc[:, [0]]
print("--- 2. Columna 0 como DATAFRAME (Lista de índices) ---")
print(columna_dataframe)
print(f"Tipo de Objeto: {type(columna_dataframe)}")
print(f"Forma (Shape): {columna_dataframe.shape}")
print("\n")


# 3. Demostración con múltiples columnas (siempre devuelve DataFrame)
# Seleccionar columnas 0, 1 y 2 (índices 0:3)
print("--- 3. Múltiples columnas (siempre DataFrame) ---")
multiples_cols = df.iloc[:, 0:3]
print(f"Tipo de Objeto: {type(multiples_cols)}")
print(f"Forma (Shape): {multiples_cols.shape}")
```

# 📍 El Método `.loc`: Localización Basada en Etiquetas

El método `.loc` de Pandas se utiliza exclusivamente para la indexación y la selección de datos **basada en etiquetas** (nombres de índice y nombres de columna).

Su sintaxis es similar a `.iloc`: **`df.loc[etiqueta_fila, etiqueta_columna]`**.

### 1. Sintaxis Básica por Etiqueta Única

| Objetivo | Código de `.loc` | Descripción |
| :--- | :--- | :--- |
| **Valor Único** | `df.loc['FilaA', 'Col_B']` | Selecciona la celda específica donde se cruzan la etiqueta de fila y columna. |
| **Fila Completa** | `df.loc['FilaA']` | Selecciona todos los valores de la fila con la etiqueta `'FilaA'`. |
| **Columna Completa** | `df.loc[:, 'Col_B']` | Selecciona todos los valores de la columna con la etiqueta `'Col_B'`. |

### 2. Selección con Listas de Etiquetas

Se pueden pasar listas de etiquetas para seleccionar múltiples filas o columnas en cualquier orden.

| Tarea | Código de `.loc` |
| :--- | :--- |
| **Filas Específicas** | `df.loc[['FilaC', 'FilaA']]` |
| **Columnas Específicas**| `df.loc[:, ['Col_A', 'Col_Z']]` |

### 3. ¡REGLA CRUCIAL!: Slicing (Rebanado) Inclusivo

A diferencia de la mayoría de las operaciones de *slicing* en Python (y de `.iloc`), cuando se usa un *slice* con `.loc`, **el elemento de la etiqueta final SÍ está incluido** en la selección.

| Slicing en Python / `.iloc` | Slicing en `.loc` |
| :--- | :--- |
| **Exclusivo:** `[inicio:fin]` excluye `fin`. | **Inclusivo:** `loc[inicio:fin]` **incluye** `fin`. |
| `[0:3]` selecciona elementos en posición 0, 1, 2. | `loc['A':'C']` selecciona filas etiquetadas como 'A', 'B', **'C'**. |

```python
import pandas as pd
import numpy as np

# 1. Crear un DataFrame de ejemplo con etiquetas de índice y columna personalizadas
data = {
    'Precio': [100, 200, 300, 400, 500],
    'Cantidad': [1, 5, 2, 8, 3],
    'Ventas': [100, 1000, 600, 3200, 1500]
}
etiquetas_filas = ['ProductoA', 'ProductoB', 'ProductoC', 'ProductoD', 'ProductoE']
df = pd.DataFrame(data, index=etiquetas_filas)

print("--- DataFrame Inicial con Etiquetas Personalizadas ---")
print(df)
print("\n" + "="*60 + "\n")


# === 1. Selección por Etiqueta Única ===

# a) Valor único (Fila 'ProductoC', Columna 'Cantidad')
valor_unico = df.loc['ProductoC', 'Cantidad']
print(f"1a. Valor único en [ProductoC, Cantidad]: {valor_unico}")


# b) Fila completa (ProductoD)
fila_completa = df.loc['ProductoD']
print("\n1b. Fila completa 'ProductoD':")
print(fila_completa)


# c) Columna completa (Precio)
columna_completa = df.loc[:, 'Precio']
print("\n1c. Columna completa 'Precio':")
print(columna_completa)
print("\n" + "="*60 + "\n")


# === 2. Selección con Listas de Etiquetas ===

# a) Filas específicas no contiguas
filas_lista = df.loc[['ProductoE', 'ProductoA', 'ProductoC']]
print("2a. Filas específicas [ProductoE, ProductoA, ProductoC]:")
print(filas_lista)


# b) Filas y columnas específicas
subconjunto = df.loc[['ProductoB', 'ProductoD'], ['Ventas', 'Precio']]
print("\n2b. Subconjunto de filas y columnas específicas:")
print(subconjunto)
print("\n" + "="*60 + "\n")


# === 3. Slicing (Rebanado) Inclusivo con Etiquetas ===

# Seleccionar desde 'ProductoA' hasta 'ProductoC' (¡'ProductoC' está incluido!)
slice_inclusivo_filas = df.loc['ProductoA':'ProductoC']
print("3a. Slicing Inclusivo (ProductoA:ProductoC):")
print(slice_inclusivo_filas)


# Slicing Inclusivo en columnas (Desde 'Precio' hasta 'Ventas')
slice_inclusivo_cols = df.loc[:, 'Precio':'Ventas']
print("\n3b. Slicing Inclusivo de Columnas (Precio:Ventas):")
print(slice_inclusivo_cols)
```

# 🎭 Filtrado Condicional (Máscaras Booleanas)

El **filtrado condicional** es la técnica más común y poderosa para seleccionar un subconjunto de filas en un DataFrame de Pandas que satisfacen una o más condiciones específicas. Esta técnica utiliza algo llamado **Máscara Booleana**.

### 1. La Máscara Booleana

Una máscara booleana es una `Series` de valores `True` (Verdadero) y `False` (Falso) que tiene la misma longitud que el DataFrame.

* Si el valor es **`True`**, la fila en esa posición será **seleccionada**.
* Si el valor es **`False`**, la fila será **descartada**.

**Creación de la Máscara:**
Se crea aplicando una condición lógica a una columna, por ejemplo: `df['housing_median_age'] > 50`

### 2. Aplicación del Filtro

Para aplicar la máscara, se coloca dentro de los corchetes de indexación (`[]`) o, de forma más segura y común, dentro de **`.loc`**:

$$\text{df.loc}[\text{Máscara Booleana}, \text{columnas_a_seleccionar}]$$

### 3. Combinando Múltiples Condiciones

Podemos combinar varias condiciones lógicas usando operadores especiales:

| Operador | Significado | Ejemplo | Descripción |
| :--- | :--- | :--- | :--- |
| **`&`** | AND (Y lógico) | `(Cond1 & Cond2)` | La fila se selecciona si **ambas** condiciones son `True`. |
| **`|`** | OR (O lógico) | `(Cond1 | Cond2)` | La fila se selecciona si **al menos una** condición es `True`. |
| **`~`** | NOT (Negación) | `(~Cond1)` | Invierte la condición (selecciona donde la condición **NO** se cumple). |

> **¡Importante!** Al combinar condiciones, **siempre** deben encerrarse las condiciones individuales entre **paréntesis `()`** para asegurar el orden de las operaciones.

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.
try:
    # Usamos una copia de las columnas relevantes para este ejemplo
    df = df_california[['median_income', 'housing_median_age', 'population']].copy()
except NameError:
    # Si no existe la variable, crea un DataFrame de prueba
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrame de prueba.")
    data = {
        'median_income': np.random.uniform(1.0, 7.0, 15),
        'housing_median_age': np.random.randint(5, 55, 15),
        'population': np.random.randint(100, 5000, 15)
    }
    df = pd.DataFrame(data)

print("--- DataFrame Inicial (Primeras 5 filas) ---")
print(df.head())
print("\n" + "="*70 + "\n")


# === 1. Filtro Condicional Básico ===

# Queremos las casas donde el ingreso medio (median_income) es mayor a 6
condicion_basica = df['median_income'] > 6
print("1a. Máscara Booleana (Primeras 5):")
print(condicion_basica.head())

# Aplicar el filtro para obtener solo las filas que cumplen la condición
filtro_alto_ingreso = df[condicion_basica]
# Nota: Usamos df[] directamente, pero df.loc[condicion_basica, :] es más explícito
print("\n1b. Filas con median_income > 6:")
print(filtro_alto_ingreso.head())
print(f"Total de filas filtradas: {len(filtro_alto_ingreso)}")
print("\n" + "="*70 + "\n")


# === 2. Combinando Múltiples Condiciones (AND / Y Lógico) ===

# Queremos las casas con:
# 1. Ingreso medio (median_income) mayor a 4
# Y
# 2. Edad media de la vivienda (housing_median_age) menor a 15 años
filtro_compuesto = (df['median_income'] > 4) & (df['housing_median_age'] < 15)

# Aplicar el filtro y seleccionar solo las columnas de 'median_income' y 'population'
resultado_and = df.loc[filtro_compuesto, ['median_income', 'population']]
print("2. Filas (Income > 4 Y Age < 15):")
print(resultado_and.head())
print(f"Total de filas filtradas: {len(resultado_and)}")
print("\n" + "="*70 + "\n")


# === 3. Usando la Negación (NOT / ~) ===

# Queremos todas las filas **EXCEPTO** aquellas cuya población sea mayor o igual a 3000
filtro_not = ~(df['population'] >= 3000)

print("3. Filas con población menor a 3000 (usando ~):")
print(df[filtro_not].head())
print(f"Total de filas filtradas: {len(df[filtro_not])}")
```

# ✍️ Creación y Modificación de Columnas

En Pandas, un DataFrame se comporta como un diccionario donde las claves son los nombres de las columnas. Para crear o modificar una columna, simplemente asignas una `Series` (una columna) o un valor único a un nombre de columna:

$$\text{df} [ '\text{nombre\_columna\_nueva}' ] = \text{valores\_o\_cálculo}$$

### 1. Creación de Columnas por Valor Único

Se asigna el mismo valor a todas las filas de la nueva columna.

* **Uso:** Inicializar una columna de bandera o categoría constante.
* **Ejemplo:** `df['Es_California'] = True`

### 2. Creación de Columnas por Cálculos

Esta es la forma más común. Se crean nuevas columnas combinando operaciones aritméticas con columnas existentes.

* **Ejemplo:** Calcular la densidad de población (población por unidad de área).
* **Fórmula:** `Habitantes_por_Hogar = df['population'] / df['households']`

### 3. Modificación de Columnas Existentes

Si el nombre de la columna ya existe en el DataFrame, la asignación sobrescribirá completamente los valores de esa columna.

* **Ejemplo:** Estandarizar una columna.
* **Fórmula:** `df['total_rooms'] = df['total_rooms'] / 1000` (convertir a miles de habitaciones)

### 4. Creación de Columnas Basada en Lógica Condicional

Para crear una columna con valores que dependen de una condición, se recomienda usar el método **`np.where()`** de NumPy por su eficiencia.

* **Sintaxis:** `np.where(condición, valor_si_True, valor_si_False)`
* **Ejemplo:** Clasificar el ingreso medio en 'Alto' o 'Bajo'.

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.
try:
    # Seleccionamos las columnas necesarias para los cálculos
    df = df_california[['population', 'households', 'total_rooms', 'median_income']].copy()
    # Reducimos el tamaño para una visualización más fácil
    df = df.head(10)
except NameError:
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrame de prueba.")
    data = {
        'population': np.random.randint(100, 5000, 10),
        'households': np.random.randint(50, 1000, 10),
        'total_rooms': np.random.randint(1000, 50000, 10),
        'median_income': np.random.uniform(1.0, 7.0, 10)
    }
    df = pd.DataFrame(data)

print("--- DataFrame Inicial (Columnas para cálculos) ---")
print(df)
print("\n" + "="*70 + "\n")


# === 1. Creación de Columna por Cálculo (Ingeniería de Características) ===

# Calcular el promedio de habitaciones por hogar
df['rooms_per_household'] = df['total_rooms'] / df['households']
print("1. Columna 'rooms_per_household' creada:")
print(df[['total_rooms', 'households', 'rooms_per_household']].head())
print("\n" + "="*70 + "\n")


# === 2. Creación de Columna por Valor Único ===

# Asignar una categoría constante 'Test_Data'
df['data_set_type'] = 'Test_Data'
print("2. Columna 'data_set_type' creada (valor único):")
print(df[['data_set_type']].head())
print("\n" + "="*70 + "\n")


# === 3. Modificación de Columna Existente ===

# Modificar 'total_rooms' para que esté expresado en miles de unidades (dividir por 1000)
df['total_rooms'] = df['total_rooms'] / 1000
print("3. Columna 'total_rooms' modificada (valores divididos por 1000):")
print(df[['total_rooms']].head())
print("\n" + "="*70 + "\n")


# === 4. Creación de Columna Basada en Lógica Condicional (np.where) ===

# Clasificar el ingreso medio (median_income) en 'Alto' o 'Bajo' (usando un umbral de 4)
df['income_level'] = np.where(
    df['median_income'] > 4,  # Condición: ¿Ingreso medio es mayor a 4?
    'Alto',                   # Valor si la condición es True
    'Bajo'                    # Valor si la condición es False
)

print("4. Columna 'income_level' creada con np.where:")
print(df[['median_income', 'income_level']])
```

# 📊 Agrupación y Agregación con `.groupby()`

El método **`.groupby()`** es fundamental para resumir y obtener estadísticas descriptivas de tu DataFrame por categorías específicas. Funciona siguiendo tres pasos clave (el proceso *Split-Apply-Combine*):

1.  **Split (Dividir):** El DataFrame se divide en grupos basándose en los valores únicos de una o más columnas categóricas que definas.
2.  **Apply (Aplicar):** Se aplica una función de agregación (como `mean()`, `sum()`, `count()`, `max()`) a cada grupo.
3.  **Combine (Combinar):** Los resultados se combinan en un nuevo DataFrame (o Serie), donde el índice es la columna de agrupación.

### 1. Sintaxis Básica

La sintaxis siempre requiere una columna para agrupar y una función para agregar:

$$\text{df}.\text{groupby}(\text{'columna\_para\_agrupar'})[\text{'columna\_para\_calcular'}] . \text{función\_agregación}()$$

### 2. Funciones de Agregación Comunes

| Función | Descripción |
| :--- | :--- |
| **`.mean()`** | Calcula el promedio (media) de los valores en cada grupo. |
| **`.sum()`** | Suma todos los valores en cada grupo. |
| **`.count()`**| Cuenta el número de valores no nulos en cada grupo. |
| **`.max()` / `.min()`**| Encuentra el valor máximo / mínimo en cada grupo. |
| **`.median()`**| Calcula la mediana (percentil 50) de cada grupo. |

### 3. Agregación de Múltiples Columnas

Para aplicar la función de agregación a **todas** las columnas numéricas restantes, simplemente omite la selección de la columna de cálculo:

* **Ejemplo:** `df.groupby('ocean_proximity').mean()`

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.
try:
    # Creamos un DataFrame de trabajo con las columnas necesarias
    df = df_california[['median_income', 'population', 'housing_median_age']].copy()

    # PASO 1: Crear una columna categórica temporal para agrupar (requerida ya que 'ocean_proximity' no existe)
    # Clasificamos la edad media de la vivienda en dos grupos: 'Antigua' (>= 30) o 'Nueva' (< 30)
    df['age_group'] = np.where(df['housing_median_age'] >= 30, 'Antigua', 'Nueva')

except NameError:
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrame de prueba.")
    data = {
        'median_income': [5.0, 6.1, 2.5, 4.0, 7.2, 3.9],
        'population': [1000, 500, 2000, 300, 800, 1500],
        'housing_median_age': [35, 15, 45, 25, 5, 30]
    }
    df = pd.DataFrame(data)
    df['age_group'] = np.where(df['housing_median_age'] >= 30, 'Antigua', 'Nueva')

print("--- DataFrame Inicial con Columna de Agrupación (age_group) ---")
print(df.head(10))
print("\n" + "="*70 + "\n")


# === 1. Agrupación Básica: Calcular la media de 'median_income' por 'age_group' ===
print("1. Ingreso Medio (mean) por Grupo de Edad:")
ingreso_promedio = df.groupby('age_group')['median_income'].mean()
print(ingreso_promedio)
print("\n" + "="*70 + "\n")


# === 2. Contar el número de filas (distritos) en cada grupo ===
print("2. Conteo (count) de Filas por Grupo de Edad:")
conteo_grupos = df.groupby('age_group')['population'].count()
print(conteo_grupos)
print("\n" + "="*70 + "\n")


# === 3. Agregación de Múltiples Columnas: Media de todas las columnas numéricas ===
print("3. Media (mean) de TODAS las Columnas Numéricas por Grupo de Edad:")
media_multiple = df.groupby('age_group').mean(numeric_only=True)
print(media_multiple)
```

# ➕ Agregación Múltiple con `.agg()` (o `.aggregate()`)

Mientras que `.mean()` o `.sum()` solo calculan una métrica, el método **`.agg()`** te permite aplicar **varias funciones de agregación** a las columnas de tus grupos en una sola llamada.

Esto es fundamental para obtener un resumen estadístico completo de cada grupo.

### 1. Sintaxis de Agregación Única

Para aplicar múltiples funciones a **una sola columna**, pasas una **lista de cadenas** (strings) con los nombres de las funciones a aplicar.

$$\text{df}.\text{groupby}('columna\_agrupacion')[\text{'columna\_calculo'}].\text{agg}( [ '\text{mean}', '\text{min}', '\text{max}' ] )$$

### 2. Sintaxis de Agregación por Columna (Diccionario)

Para aplicar **diferentes** funciones de agregación a **diferentes** columnas al mismo tiempo, pasas un **diccionario** a `.agg()`:

* **Clave del diccionario:** El nombre de la columna a la que se le aplicará la función.
* **Valor del diccionario:** Una cadena o una lista de cadenas con las funciones de agregación.

| Columna a Calcular | Función/es a Aplicar |
| :--- | :--- |
| `'median_income'` | `['mean', 'median']` |
| `'population'` | `'sum'` |

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.
try:
    # Creamos un DataFrame de trabajo con las columnas necesarias
    df = df_california[['median_income', 'population', 'housing_median_age']].copy()

    # Creamos la columna categórica 'age_group' para agrupar (como en el ejemplo anterior)
    df['age_group'] = np.where(df['housing_median_age'] >= 30, 'Antigua', 'Nueva')

except NameError:
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrame de prueba.")
    data = {
        'median_income': [5.0, 6.1, 2.5, 4.0, 7.2, 3.9],
        'population': [1000, 500, 2000, 300, 800, 1500],
        'housing_median_age': [35, 15, 45, 25, 5, 30]
    }
    df = pd.DataFrame(data)
    df['age_group'] = np.where(df['housing_median_age'] >= 30, 'Antigua', 'Nueva')

print("--- DataFrame Inicial con Columna de Agrupación (age_group) ---")
print(df.head())
print("\n" + "="*70 + "\n")


# === 1. Agregación Múltiple a UNA Columna (usando lista) ===

# Calcular la media, el mínimo y el máximo de 'median_income' por 'age_group'
resumen_ingreso = df.groupby('age_group')['median_income'].agg(['mean', 'min', 'max'])
print("1. Resumen de 'median_income' por Grupo de Edad (mean, min, max):")
print(resumen_ingreso)
print("\n" + "="*70 + "\n")


# === 2. Agregación Múltiple a MÚLTIPLES Columnas (usando diccionario) ===

# Aplicamos diferentes funciones a diferentes columnas:
# - 'median_income': Calcular media y mediana
# - 'population': Calcular suma y máximo
resumen_completo = df.groupby('age_group').agg(
    {
        'median_income': ['mean', 'median'],
        'population': ['sum', 'max']
    }
)
print("2. Resumen Completo (Agregación diferente por columna):")
print(resumen_completo)

# Nota: El resultado tiene un MultiIndex en las columnas, para acceder a 'sum' de 'population':
# print("\nPoblación Total de Grupos Antiguos:", resumen_completo.loc['Antigua', ('population', 'sum')])
```

# ⬇️ Ordenamiento de Datos con `.sort_values()`

El método **`.sort_values()`** se utiliza para ordenar las filas de un DataFrame basándose en los valores contenidos en una o más columnas.

### 1. Sintaxis Básica

Necesitas especificar la columna por la cual quieres ordenar y la dirección del ordenamiento.

$$\text{df}.\text{sort\_values}(\text{by} = '\text{columna}', \text{ascending} = \text{True/False})$$

| Argumento | Descripción | Valores |
| :--- | :--- | :--- |
| **`by`** | **Obligatorio.** Nombre de la columna (o lista de columnas) por la que se ordenará. | Cadena o Lista de cadenas. |
| **`ascending`**| **Opcional.** Define la dirección del ordenamiento. | `True` (predeterminado: Ascendente, de menor a mayor) o `False` (Descendente, de mayor a menor). |

### 2. Ordenamiento por Múltiples Columnas

Puedes ordenar por varias columnas a la vez, lo cual es útil para desempatar. Pandas ordena primero por la primera columna de la lista, y si hay valores iguales, usa la segunda, y así sucesivamente.

| Tarea | Código Python |
| :--- | :--- |
| **Ordenamiento Simple** | `df.sort_values(by='median_income', ascending=False)` |
| **Ordenamiento Múltiple** | `df.sort_values(by=['housing_median_age', 'population'], ascending=[True, False])` |

> **Nota:** Cuando usas ordenamiento múltiple, puedes especificar una dirección (`ascending`) para cada columna de la lista. En el ejemplo anterior, ordena la edad de forma ascendente y la población de forma descendente.

### 3. Modificación en el Lugar (`inplace=True`)

Por defecto, `.sort_values()` devuelve un nuevo DataFrame ordenado. Si quieres que la modificación se aplique directamente al DataFrame original (`df`), debes usar el argumento `inplace=True`.

* **Ejemplo:** `df.sort_values(by='median_income', ascending=True, inplace=True)`

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.
try:
    # Creamos un DataFrame de trabajo con las columnas necesarias
    df = df_california[['median_income', 'population', 'housing_median_age']].copy()
    df = df.head(15)

except NameError:
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrame de prueba.")
    data = {
        'median_income': [5.0, 6.1, 2.5, 4.0, 7.2, 3.9, 5.0, 2.0, 6.1, 1.5],
        'population': [1000, 500, 2000, 300, 800, 1500, 1000, 400, 500, 100],
        'housing_median_age': [35, 15, 45, 25, 5, 30, 35, 50, 15, 40]
    }
    df = pd.DataFrame(data)

print("--- DataFrame Inicial (No Ordenado) ---")
print(df)
print("\n" + "="*70 + "\n")


# === 1. Ordenamiento Simple Descendente (Mayor a Menor) ===

# Encontrar los distritos con el ingreso medio más alto
df_ordenado_ingreso = df.sort_values(by='median_income', ascending=False)
print("1. Ordenado por 'median_income' DESCENDENTE (Más ricos primero):")
print(df_ordenado_ingreso.head())
print("\n" + "="*70 + "\n")


# === 2. Ordenamiento Múltiple y Direcciones Mixtas ===

# Ordenar primero por 'housing_median_age' ascendente (más joven primero)
# Luego, si las edades son iguales, ordenar por 'population' descendente (más poblado primero)
df_ordenado_doble = df.sort_values(
    by=['housing_median_age', 'population'],
    ascending=[True, False]
)
print("2. Ordenado por Age (ASC) y Population (DESC):")
print(df_ordenado_doble)
```

# 🗑️ Manejo de Valores Faltantes (NaN)

Los valores faltantes (Not a Number, `NaN`) representan datos ausentes o desconocidos. Pandas proporciona métodos robustos para localizarlos, eliminarlos o imputarlos (rellenarlos).

### 1. Identificación y Conteo

Antes de actuar, es fundamental saber dónde y cuántos valores faltantes hay.

| Tarea | Código Python |
| :--- | :--- |
| **Identificar Nulos** | `df.isnull()` | Devuelve un DataFrame Booleano (`True` si es nulo). |
| **Contar Nulos (Columna)**| `df['columna'].isnull().sum()`| Cuenta el total de nulos en la columna. |
| **Contar Nulos (Todo DF)** | `df.isnull().sum()` | Devuelve el conteo de nulos para cada columna. |

### 2. Eliminación de Filas/Columnas (`.dropna()`)

El método **`.dropna()`** permite eliminar filas o columnas enteras si contienen algún valor `NaN`.

| Argumento | Descripción | Valores |
| :--- | :--- | :--- |
| **`axis`** | Define si se eliminan filas (`0` o `'index'`) o columnas (`1` o `'columns'`). | `0` (predeterminado) o `1`. |
| **`how`** | Define la regla de eliminación. | `any` (elimina si **algún** valor es `NaN`) o `all` (elimina si **todos** los valores son `NaN`). |

### 3. Relleno de Valores Faltantes (`.fillna()`)

El método **`.fillna()`** sustituye los valores `NaN` por un valor específico o por un método de imputación.

| Tarea | Código Python | Descripción |
| :--- | :--- | :--- |
| **Rellenar con Valor Único** | `df['col'].fillna(0)` | Sustituye todos los `NaN` por cero. |
| **Rellenar con la Media** | `df['col'].fillna(df['col'].mean())`| Sustituye `NaN` por el promedio de la columna. |
| **Rellenar con Mediana** | `df['col'].fillna(df['col'].median())`| Sustituye `NaN` por la mediana de la columna (más robusta a *outliers*). |

> **Nota:** Al igual que con `.sort_values()`, tanto `.dropna()` como `.fillna()` devuelven un nuevo DataFrame. Para modificar el original, usa `inplace=True`.

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.
try:
    # Usamos las columnas relevantes y creamos una copia
    df = df_california[['total_bedrooms', 'population', 'median_income']].copy()

    # Aseguramos la existencia de NaN: total_bedrooms tiene algunos NaN reales.
    # Forzamos NaN en otras columnas para el ejemplo:
    df.loc[10:12, 'population'] = np.nan
    df.loc[5, 'median_income'] = np.nan

except NameError:
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrame de prueba.")
    data = {
        'total_bedrooms': [400.0, np.nan, 250.0, 300.0, np.nan],
        'population': [1000, 500, np.nan, 300, 800],
        'median_income': [5.0, 6.1, 2.5, np.nan, 7.2]
    }
    df = pd.DataFrame(data)

print("--- DataFrame Inicial con Valores NaN ---")
print(df.head(15))
print("\n" + "="*70 + "\n")


# === 1. Identificación y Conteo de Nulos ===
print("1a. Conteo de Nulos por Columna (df.isnull().sum()):")
print(df.isnull().sum())
print("\n")

# 1b. Conteo del DataFrame completo
nulos_totales = df.isnull().sum().sum()
print(f"1b. Total de valores nulos en el DataFrame: {nulos_totales}")
print("\n" + "="*70 + "\n")


# === 2. Eliminación de Filas con .dropna() ===

# Crear un DataFrame limpio eliminando cualquier fila con al menos un NaN
df_sin_nulos = df.dropna(how='any', axis=0)
print("2. Filas restantes después de eliminar cualquier fila con un NaN (df.dropna()):")
print(df_sin_nulos.head())
print(f"Filas originales: {len(df)} | Filas restantes: {len(df_sin_nulos)}")
print("\n" + "="*70 + "\n")


# === 3. Relleno de Valores con .fillna() ===

# Rellenar 'total_bedrooms' con el valor de la media de esa columna
media_bedrooms = df['total_bedrooms'].mean()
print(f"3a. Media de total_bedrooms: {media_bedrooms:.2f}")

# Aplicar el relleno solo a 'total_bedrooms'
df['total_bedrooms_filled'] = df['total_bedrooms'].fillna(media_bedrooms)

# Rellenar 'population' con un valor constante (ej. 0)
df['population_filled'] = df['population'].fillna(0)

# Comparación del relleno (muestra las filas donde originalmente había NaN)
print("\n3b. Comparación de Columnas Rellenadas (Media y Constante):")
print(df[df['total_bedrooms'].isnull() | df['population'].isnull()][
    ['total_bedrooms', 'total_bedrooms_filled', 'population', 'population_filled']
])
```

# 🧩 Combinación de DataFrames: `merge` y `concat`

Combinar DataFrames significa unir dos o más tablas en una sola. Los dos métodos principales tienen propósitos distintos:

### 1. `pd.merge()` (JOIN por Clave)

Se utiliza para combinar DataFrames **horizontalmente** basándose en valores comunes en una o varias columnas clave (similar a las operaciones JOIN en SQL).

| Argumento | Descripción | Valores Típicos |
| :--- | :--- | :--- |
| **`on`** | Columna(s) clave para la unión. | `'columna_id'` o `['col1', 'col2']` |
| **`how`** | Tipo de JOIN a realizar. | `'inner'`, `'left'`, `'right'`, `'outer'` |

#### Tipos Comunes de `how` (JOIN)

* **`inner` (Predeterminado):** Solo incluye las filas que tienen coincidencias en **ambos** DataFrames.
* **`left`:** Incluye todas las filas del DataFrame de la izquierda (`df1`) y solo las que coinciden del de la derecha (`df2`). Rellena con `NaN` si no hay coincidencia.
* **`outer`:** Incluye **todas** las filas de ambos DataFrames. Rellena con `NaN` donde no haya coincidencias.

[Image of pd.merge types inner left right outer join]

### 2. `pd.concat()` (Concatenación/Apilamiento)

Se utiliza para apilar DataFrames **verticalmente** (más filas) u **horizontalmente** (más columnas), generalmente cuando tienen la misma estructura o necesitan alinearse por índice.

| Argumento | Descripción | Valores Típicos |
| :--- | :--- | :--- |
| **`axis`** | Dirección de la unión. | `0` (predeterminado: apilar filas, vertical) o `1` (apilar columnas, horizontal). |
| **`join`** | Cómo manejar las columnas/filas diferentes. | `'outer'` (predeterminado: incluye todas las columnas) o `'inner'` (solo incluye columnas comunes). |

```python
import pandas as pd
import numpy as np

# AVISO: El código asume que la variable df_california ya está cargada.

# --- 1. PREPARACIÓN DE DATOS DE EJEMPLO ---

# Usaremos df_california para crear dos DataFrames de ejemplo (df1 y df2)
try:
    df_base = df_california[['longitude', 'latitude', 'median_income']].head(10).copy()

    # Crear un DF de 'metricas_adicionales' que contiene la misma longitud, latitud y una nueva columna
    metricas_adicionales = df_california[['longitude', 'latitude', 'housing_median_age']].head(10).copy()
    metricas_adicionales['age_category'] = np.where(metricas_adicionales['housing_median_age'] > 30, 'Antigua', 'Nueva')

    # Simular un DF con menos filas (para ver el efecto del JOIN)
    metricas_adicionales = metricas_adicionales.iloc[0:7]

except NameError:
    print("ADVERTENCIA: df_california no encontrado. Creando DataFrames de prueba.")
    df_base = pd.DataFrame({'id': [1, 2, 3, 4], 'valor_a': [10, 20, 30, 40]})
    metricas_adicionales = pd.DataFrame({'id': [1, 3, 5], 'valor_b': [100, 300, 500]})

# Renombramos las columnas en el ejemplo de prueba para usar 'id' como clave
if 'id' in df_base.columns:
    df1 = df_base
    df2 = metricas_adicionales
    key_col = 'id'
else:
    # Si estamos en el caso real de california_housing, usamos longitud/latitud como clave de unión
    df1 = df_base
    df2 = metricas_adicionales
    key_col = ['longitude', 'latitude']


print("--- DF Izquierda (DF1) ---")
print(df1)
print("\n--- DF Derecha (DF2) ---")
print(df2)
print("\n" + "="*70 + "\n")


# === 1. Combinación Horizontal con pd.merge() ===

# Realizamos un LEFT JOIN: Mantenemos todas las filas de DF1 y solo añadimos las que coincidan de DF2
merge_left = pd.merge(df1, df2, on=key_col, how='left')
print("1a. Resultado de pd.merge(how='left'):")
print(merge_left)


# Realizamos un INNER JOIN: Solo filas presentes en ambos (debería ser más corto si hay IDs faltantes)
merge_inner = pd.merge(df1, df2, on=key_col, how='inner')
print("\n1b. Resultado de pd.merge(how='inner'):")
print(merge_inner)
print("\n" + "="*70 + "\n")


# === 2. Concatenación Vertical con pd.concat() ===

# Crear un tercer DF con la misma estructura para concatenar
df_extra = df1.head(2).copy()
df_extra['median_income'] = df_extra['median_income'] * 2
print("--- DF Extra para Concatenar ---")
print(df_extra)

# Concatenar DF1 y DF_Extra (apilar uno debajo del otro, axis=0 por defecto)
df_concat_vertical = pd.concat([df1, df_extra], ignore_index=True)
print("\n2. Resultado de pd.concat() vertical (Más filas):")
print(df_concat_vertical)
```
